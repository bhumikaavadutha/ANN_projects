```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Activation functions and their derivatives
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

def leaky_relu_derivative(x, alpha=0.01):
    dx = np.ones_like(x)
    dx[x < 0] = alpha
    return dx

def softmax(x):
    exps = np.exp(x - np.max(x))  # subtracting max(x) for numerical stability
    return exps / np.sum(exps)

def softmax_derivative(x):
    s = softmax(x)
    return s * (1 - s)
```

```python
# Range for x-axis
x = np.linspace(-5, 5, 100)
x
```

```python
# Range for x-axis
x = np.linspace(-5, 5, 100)

# Plotting activation functions
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.title('Sigmoid Function')
plt.plot(x, sigmoid(x), label='sigmoid')
plt.plot(x, sigmoid_derivative(x), label='sigmoid derivative')
plt.legend()
```

```python
plt.subplot(2, 2, 1)
plt.title('Hyperbolic Tangent (Tanh) Function')
plt.plot(x, tanh(x), label='tanh')
plt.plot(x, tanh_derivative(x), label='tanh derivative')
plt.legend()

```

```python
plt.subplot(2, 2, 3)
plt.title('ReLU Function')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, relu_derivative(x), label='ReLU derivative')
plt.legend()
```

```python

plt.subplot(2, 2, 4)
plt.title('Leaky ReLU Function')
plt.plot(x, leaky_relu(x), label='Leaky ReLU')
plt.plot(x, leaky_relu_derivative(x), label='Leaky ReLU derivative')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 4)
plt.title('Softmax Function')
plt.plot(x, softmax(x), label='Softmax')
plt.plot(x, softmax_derivative(x), label='Softmax derivative')
plt.legend()
```
