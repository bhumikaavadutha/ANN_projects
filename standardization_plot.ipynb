```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load the data from the CSV file
data = pd.read_csv('/home/bhumika-avadutha/Downloads/archive(1)/CompanyABCProfit.csv')  # Replace 'your_data.csv' with your file name

# Assuming the column of interest is named 'value'
data_column = data['Year']  # Replace 'value' with the name of the column in your CSV file

# Calculate the mean and standard deviation of the data
mean = np.mean(data_column)
std_dev = np.std(data_column)


# Plot the histogram of the data
plt.figure(figsize=(10, 6))
plt.hist(data_column, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Data Histogram')

# Create the range for the x-axis
x = np.linspace(min(data_column), max(data_column), 1000)

# Calculate the PDF of the normal distribution for the given mean and standard deviation
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Plot the PDF on the same plot
plt.plot(x, pdf, 'r-', label=f'Normal Distribution\nmean={mean:.2f}, std={std_dev:.2f}')

# Add labels and title
plt.title('Histogram of Data and Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
